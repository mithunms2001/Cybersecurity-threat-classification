```python
import os
import pandas as pd
import numpy as np

def preprocess_data(df, output_file_path, missing_threshold=0.5):
    try:
        # Ensure the output directory exists
        output_dir = os.path.dirname(output_file_path)
        os.makedirs(output_dir, exist_ok=True)
        
        # Load the DataFrame
        print(f"Loaded data with {df.shape[0]} rows and {df.shape[1]} columns.")

        # Drop columns with high missing values
        missing_percentages = df.isnull().mean()
        columns_to_drop = missing_percentages[missing_percentages > missing_threshold].index
        df.drop(columns=columns_to_drop, inplace=True)
        print(f"Columns dropped due to more than {missing_threshold * 100}% missing values: {columns_to_drop.tolist()}")

        # Remove duplicates
        df.drop_duplicates(inplace=True)
        print(f"Duplicates removed. Number of rows after deduplication: {df.shape[0]}")

        # Feature extraction from 'Timestamp'
        if 'Timestamp' in df.columns:
            df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
            df['Year'] = df['Timestamp'].dt.year
            df['Month'] = df['Timestamp'].dt.month
            df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
            df['Hour'] = df['Timestamp'].dt.hour
            df.drop(columns=['Timestamp'], inplace=True)
            print("Timestamp features extracted.")

        # Fill missing values in specific columns with their mode
        for column in ['CountryCode', 'State', 'City']:
            if column in df.columns:
                mode_value = df[column].mode()[0]
                df[column].fillna(mode_value, inplace=True)
                print(f"Filled missing values in '{column}' with mode: {mode_value}")

        # Fill missing values in 'IncidentGrade' equally
        if 'IncidentGrade' in df.columns:
            original_value_counts = df['IncidentGrade'].value_counts()
            print("Original value counts before filling missing values:")
            print(original_value_counts)

            unique_values = df['IncidentGrade'].dropna().unique()
            num_missing = df['IncidentGrade'].isnull().sum()
            fill_values = [value for value in unique_values for _ in range(num_missing // len(unique_values))]
            remaining = num_missing % len(unique_values)
            fill_values.extend(np.random.choice(unique_values, remaining))
            np.random.shuffle(fill_values)
            df.loc[df['IncidentGrade'].isnull(), 'IncidentGrade'] = fill_values
            filled_value_counts = df['IncidentGrade'].value_counts()
            print("\nValue counts after filling missing values:")
            print(filled_value_counts)

        # Convert all columns to 'category' dtype
        df = df.astype('category')

        # Save the preprocessed DataFrame
        df.to_csv(output_file_path, index=False)
        print(f"Preprocessed data saved to {output_file_path}.")

    except Exception as e:
        print(f"An error occurred: {e}")

# Specify the output file path
output_file_path = r"C:/Users/Home/Cybersecurity/preprocessed_data_1.csv"  

# Run the preprocessing function
preprocess_data(df, output_file_path)
```

```python
import pandas as pd

# Read the CSV file
file_path = r"C:/Users/Home/Cybersecurity/top_n_categories_1.csv"
df = pd.read_csv(file_path)

# Display initial DataFrame
print("Initial DataFrame:", df.head())

# Function to identify and remove outliers
def identify_and_remove_outliers(df):
    outlier_indices = []
    for col in df.select_dtypes(include=[float, int]).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_indices.extend(outliers.index.tolist())
    
    outlier_indices = list(set(outlier_indices))
    outlier_rows = df.loc[outlier_indices]
    df_cleaned = df.drop(outlier_indices)
    
    return df_cleaned, outlier_rows

# Identify and remove outliers
df_cleaned, outlier_rows = identify_and_remove_outliers(df)

# Print the rows identified as outliers
print("\nRows identified as outliers:", outlier_rows)

# Print the cleaned DataFrame
print("\nCleaned DataFrame:", df_cleaned.head())

# Print the rows after removing outliers
print("\nRows after removing outliers:", df_cleaned)

# Save the cleaned DataFrame to a new CSV file in the specified location
output_file_path = r"C:/Users/Home/Cybersecurity/top_n_categories_cleaned_1.csv"
df_cleaned.to_csv(output_file_path, index=False)
print(f"\nCleaned data saved to {output_file_path}")
```

```python
import pandas as pd

# Define file paths
input_file_path = r"C:/Users/Home/Cybersecurity/top_n_categories_cleaned_1.csv"
output_file_path = r"C:/Users/Home/Cybersecurity/top_n_categories_encoded_1.csv"  # Updated file path

# Define the number of top categories to retain and the replacement value for others
top_n_categories = 5
replacement_value = 6

# Function to encode top categories and replace others
def encode_categories(df, top_n=5, replacement_val=6):
    # Convert all integer columns to categorical
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        df[col] = df[col].astype('category')
    
    # Apply the encoding to each categorical column
    for col in df.select_dtypes(include=['category', 'object']).columns:
        # Convert to category type if not already
        df[col] = df[col].astype('category')
        category_counts = df[col].value_counts()
        top_categories = list(category_counts.nlargest(top_n).index)
        
        # Add 'Other' to the list of categories if not already present
        if 'Other' not in top_categories:
            top_categories.append('Other')
        
        # Ensure all categories are strings for consistent sorting and mapping
        top_categories = [str(cat) for cat in top_categories]
        
         # Create a manual mapping for label encoding, ensuring continuous integers
        unique_categories = sorted(top_categories)
        if 'Other' in unique_categories:
            unique_categories.remove('Other')
            unique_categories.append('Other')
        
        # Create the manual mapping
        manual_mapping = {category: i for i, category in enumerate(unique_categories)}
        
        # Add 'Other' to the Categorical type
        df[col] = df[col].cat.add_categories(['Other'])
        
        # Replace categories not in top_n with 'Other'
        df[col] = df[col].apply(lambda x: str(x) if str(x) in top_categories else 'Other')
        
        # Apply manual label encoding
        df[col] = df[col].map(manual_mapping)
        
        # Fill NaN values with the replacement value
        df[col] = df[col].fillna(manual_mapping['Other']).astype(int)
    
    return df

# Process the CSV file in chunks
chunk_size = 500000  # Adjust the chunk size as needed
chunks = []

for chunk in pd.read_csv(input_file_path, chunksize=chunk_size, low_memory=False):
    chunk_encoded = encode_categories(chunk, top_n=top_n_categories, replacement_val=replacement_value)
    chunks.append(chunk_encoded)

# Concatenate all chunks
df_encoded = pd.concat(chunks, axis=0)

# Save the processed DataFrame with encoded categories
df_encoded.to_csv(output_file_path, index=False)  # Saving to the updated file path

# Display the processed DataFrame
print("\nProcessed DataFrame with encoded categories:")
print(df_encoded.head())
print(f"Shape of processed DataFrame: {df_encoded.shape}")

# Display the number of unique values per column after encoding
print("\nNumber of unique values per column after encoding:")
for col in df_encoded.columns:
    num_unique_values = df_encoded[col].nunique()
    print(f"{col}: {num_unique_values}")

# Check unique values in 'IncidentGrade'
unique_values = df_encoded['IncidentGrade'].unique()
print("\nUnique values in 'IncidentGrade':", unique_values)
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2

# Read the CSV file
file_path = r"C:/Users/Home/Cybersecurity/top_n_categories_encoded_1.csv"
df = pd.read_csv(file_path)

# Define the features (X) and the target (y)
X = df.drop(columns=['IncidentGrade'])
y = df['IncidentGrade']

# Perform Chi-Square test
chi2_selector = SelectKBest(chi2, k='all')
chi2_selector.fit(X, y)

# Create DataFrame to view scores and p-values
chi2_scores = pd.DataFrame({
    'Feature': X.columns,
    'Chi2 Score': chi2_selector.scores_,
    'P-value': chi2_selector.pvalues_
})

# Sort by Chi2 Score
chi2_scores = chi2_scores.sort_values(by='Chi2 Score', ascending=False)

# Print Chi2 Scores and P-values
print(chi2_scores)

# Select features based on Chi-Square p-values
selected_features_chi = chi2_scores[chi2_scores['P-value'] < 0.05]['Feature'].tolist()
print("Selected Features:", selected_features_chi)

# Create a DataFrame with the selected features and include the target variable
df_selected_features = df[selected_features_chi + ['IncidentGrade']]

# Save the selected features to a new CSV file
output_file_path = r"C:/Users/Home/Cybersecurity/selected_features_training_1.csv"  # Updated file path
df_selected_features.to_csv(output_file_path, index=False)
print(f"Selected features saved to {output_file_path}")
```

```python
import pandas as pd
import numpy as np

# Define functions
def drop_usage_column(df):
    if 'Usage' in df.columns:
        df.drop(columns=['Usage'], inplace=True)
        print("Dropped 'Usage' column.")
    else:
        print("'Usage' column not found in the dataset.")
    return df

def drop_high_missing_values(df, threshold=0.5):
    missing_percentages = df.isnull().mean()
    columns_to_drop = missing_percentages[missing_percentages > threshold].index
    df.drop(columns=columns_to_drop, inplace=True)
    print(f"Columns dropped due to more than {threshold * 100}% missing values: {columns_to_drop.tolist()}")
    return df

def remove_duplicates(df):
    df.drop_duplicates(inplace=True)
    print(f"Duplicates removed. Number of rows after deduplication: {df.shape[0]}")
    return df

def extract_features(df):
    if 'Timestamp' in df.columns:
        df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
        df['Year'] = df['Timestamp'].dt.year
        df['Month'] = df['Timestamp'].dt.month
        df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
        df['Hour'] = df['Timestamp'].dt.hour
        df.drop(columns=['Timestamp'], inplace=True)
        print("Timestamp features extracted.")
    return df

def convert_to_category(df):
    return df.astype('category')

def replace_rare_categories(df, top_n=5, replace_name='other'):
    for col in df.select_dtypes(include=['category']).columns:
        top_categories = df[col].value_counts().nlargest(top_n).index
        df[col] = df[col].apply(lambda x: x if x in top_categories else replace_name)
        df[col] = df[col].astype('category')
    print(f"Replaced rare categories in each column with '{replace_name}'.")

def identify_and_remove_outliers(df):
    outlier_indices = []
    for col in df.select_dtypes(include=[float, int]).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_indices.extend(outliers.index.tolist())
    
    outlier_indices = list(set(outlier_indices))
    outlier_rows = df.loc[outlier_indices]
    df_cleaned = df.drop(outlier_indices)
    
    return df_cleaned, outlier_rows

# Load data
file_path = r"C:/Users/Home/Cybersecurity/GUIDE_Test.csv"
try:
    df = pd.read_csv(file_path)
    print(f"Loaded data with {df.shape[0]} rows and {df.shape[1]} columns.")
except FileNotFoundError:
    print(f"Error: The file at {file_path} was not found.")
    raise
except pd.errors.EmptyDataError:
    print("Error: The file is empty.")
    raise

# Step 1: Drop 'Usage' column
df = drop_usage_column(df)

# Step 2: Drop columns with high missing values
df = drop_high_missing_values(df)

# Step 3: Remove duplicate rows
df = remove_duplicates(df)

# Step 4: Extract features from 'Timestamp' column
df = extract_features(df)

# Step 5: Drop specified columns
columns_to_drop = [
    'Id', 'IncidentId', 'AlertId', 'DeviceId', 'Sha256', 'IpAddress', 'Url',
    'AccountSid', 'AccountUpn', 'AccountObjectId', 'AccountName', 'DeviceName',
    'NetworkMessageId', 'RegistryKey', 'RegistryValueName', 'RegistryValueData',
    'ApplicationId', 'OAuthApplicationId', 'FileName', 'FolderPath', 'ResourceIdName'
]
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
print(f"Dropped columns: {columns_to_drop}")

# Step 6: Convert all columns to 'category' dtype
df = convert_to_category(df)
print("Converted all columns to 'category' dtype.")

# Step 7: Replace rare categories
top_n_categories = 5  # Adjust the number of top categories as needed
replace_rare_categories(df, top_n=top_n_categories)

# Step 8: Identify and remove outliers
# Print the DataFrame before removing outliers
print("\nDataFrame before removing outliers:", df.head())

df_cleaned, outlier_rows = identify_and_remove_outliers(df)

# Step 9: Print the rows identified as outliers
print("\nRows identified as outliers:", outlier_rows)

# Print the DataFrame after removing outliers
print("\nDataFrame after removing outliers:", df_cleaned.head())

# Step 10: Save the cleaned DataFrame to a new CSV file
output_file_path = r"C:/Users/Home/Cybersecurity/top_n_categories_cleaned_testing_1.csv" 
df_cleaned.to_csv(output_file_path, index=False)
print(f"\nCleaned data saved to {output_file_path}")
```
